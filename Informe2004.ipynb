```python
n = set(x['Nombre Rio'])
n
```

```python
import statistics
for y in n:
    z = x[x['Nombre Rio'] == y]['Aportes Energia kWh']
    if len(z) > 0:
        print(y, statistics.mean(x[x['Nombre Rio'] == y]['Aportes Energia kWh'].values[:]))
```
